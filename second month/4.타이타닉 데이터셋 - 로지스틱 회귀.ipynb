```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#데이터셋 로드
train = pd.read_csv('D:\\AI_1400_kys\\AI2\\workspace\\day03\\train.csv')
test = pd.read_csv('D:\\AI_1400_kys\\AI2\\workspace\\day03\\test.csv')
```

```python
train
```

```python
test = test.drop(['PassengerId', 'Ticket', 'Name', 'Cabin'], axis = 1)
test
```

```python
train = train.drop('Cabin', axis = 1)
train
```

## train 컬럼별 null 값 확인
- Age, Embarked

```python
train.isnull().sum()
```

```python
#Age는 평균 나이로 Embarked N으로 채우기
train['Age'].fillna(train['Age'].mean(), inplace = True)
train['Embarked'].fillna('N', inplace = True)

train

train.isnull().sum()
```

```python
train
```

```python
test
```

```python
#성별변수를 이진변수로 변환(문자열 -> 숫자로 변환)
train['Sex'] = train['Sex'].apply(lambda x : 1 if x == 'male' else 0)
test['Sex'] = test['Sex'].apply(lambda x : 1 if x == 'male' else 0)
```

```python
train
```

```python
test
```

```python
#상관관계 확인하기 위해서 문자열 값 삭제
train = train.drop(['Name', 'Ticket', 'Embarked'], axis = 1 )
```

# 그래프로 확인

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#상관관계 그래프 확인
plt.figure(figsize = (12, 8))
sns.heatmap(train.corr(), annot = True)
```

```python
train.shape
```

```python
#범주형 변수 더피화
#.get_dummies() 범주형 변수들에 대해 더미 변수를 생성함, 범주형변수가 이진 변수로 변환됨
train = pd.get_dummies(train)
train
test = pd.get_dummies(test)
test
```

```python
X_train = train.drop(['Survived'], axis = 1)
y_train = train['Survived']
```

```python
X_train
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#모델 생성
lr = LogisticRegression()

#모델 학습
lr.fit(X_train, y_train)
```

```python
#예측
pred = lr.predict(X_train)

#정확도평가
acc = accuracy_score(y_train, pred)
print(f"정확도 : {acc}")
```
